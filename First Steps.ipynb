# Primeiros Passos em Machine Learning

```python
import pandas as pd
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score
```

```python
# 1 = sim; 0 = não
# pelo longo?
# perna curta?
# faz auau?

porco1 = [0, 1, 0]
porco2 = [0, 1, 1]
porco3 = [1, 1, 0]

cachorro1 = [0, 1, 1]
cachorro2 = [1, 0, 1]
cachorro3 = [1, 1, 1]

dados = [porco1, porco2, porco3, cachorro1, cachorro2, cachorro3] # treino_x
classes = [1, 1, 1, 0, 0, 0] # 1 = porco, 0 = cachorro            # treino y
```

```python
model = LinearSVC()
model.fit(dados, classes) # dados, gabarito
```

```python
animal_misterioso = [1, 1, 1]
model.predict([animal_misterioso]) # Lista com um item, normalmente se passariam mais itens
```

```python
misterio1 = [1, 1, 1]
misterio2 = [1, 1, 0]
misterio3 = [0, 1, 1]

testes = [misterio1, misterio2, misterio3]           # teste_x
previsoes = model.predict(testes)
previsoes
```

#### Calculando a taxa de acerto?

```python
testes_classes = [0, 1, 1] # "Gabarito"             # teste y
previsoes == testes_classes # Pode ser feito já que o sklearn usa o Numpy
```

```python
(previsoes == testes_classes).sum()
```

```python
corretos = (previsoes == testes_classes).sum()
total = len(testes)
percentual_de_acerto = corretos * 100/total
print(f'A taxa de acerto é {percentual_de_acerto:.2f} %')
```

```python
# forma dois (muito mais fácil)
percentual_de_acerto = accuracy_score(testes_classes, previsoes) * 100
print(f'A taxa de acerto é {percentual_de_acerto:.2f} %')
```
